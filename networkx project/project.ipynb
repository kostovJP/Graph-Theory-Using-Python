```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
from itertools import combinations
G = nx.Graph() #-->creates an undirected graph
#for directed graphs we can use: nx.DiGraph()
cities = ['Delhi','Bangalore','Hyderabad','Ahmedabad','Chennai','Kolkata','Surat','Pune','Jaipur']
#create a node for each city:
for c in cities:
    G.add_node(c)
G.nodes()
max_nodes = len(list(combinations(cities,2)))
print(max_nodes)
costs = [random.randint(100,2000) for i in range(max_nodes)]
print(costs)

#let us add 16 edges to this network:
while(G.number_of_edges() < 16):
    c1 = random.choice(list(G.nodes()))
    c2 = random.choice(list(G.nodes()))
    if c1 != c2 and G.has_edge(c1,c2) == 0:
        w = random.choice(costs)
        G.add_edge(c1,c2, weight = w)
G.edges()
style = nx.circular_layout(G)
nx.draw(G, style, with_labels = 1)#--->draws without the edge weights
nx.draw_networkx_edge_labels(G,style)#-->draws with the edge weights
plt.show()
```
